```python
import numpy as np
import pandas as pd
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df = pd.read_csv("real_estate.csv")
df.head()
```

```python
print(list(df.columns))
```

```python
data = df.loc[:, ['X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores', 'Y house price of unit area']]
data
```

```python
X = data.drop('Y house price of unit area', axis = 1)
X
```

```python
y = data['Y house price of unit area']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Выведите на экран коэффициенты, которые выбрала регрессионная модель
coef = pd.DataFrame(model.coef_, X.columns, columns = ['Coeficient'])
coef
```

```python
# теперь сделаем прогноз на тестовых данных
y_pred = model.predict(X_test)
```

```python
# создайте таблицу, в которой будет отображена разница между фактическим значением и прогнозируемым значением.
d = pd.DataFrame({'Act': y_test, 'Pred': y_pred})
d
```

```python
# Теперь оценим производительность алгоритма
# mean squared error (mse)
print("MSE: ", metrics.mean_squared_error(y_test, y_pred))
```
